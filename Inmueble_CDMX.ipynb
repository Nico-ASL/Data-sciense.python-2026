```python
datos['Condominio'] = datos['Condominio'].fillna(0)
```

```python
datos = datos[~((datos['Tipo'] == 'Departamento') & (datos['Valor'] == 0))]
```

```python
datos = datos[~((datos['Tipo'] == 'Departamento') & (datos['Condominio'] == 0))]
```

```python
filtro1 = (datos['Tipo'] == 'Departamento') & (datos['Habitaciones'] == 1) & (datos['Valor'] < 4200)
```

```python
filtro2 = (datos['Tipo'] == 'Departamento') & (datos['Habitaciones'] >= 2) & (datos['Valor'] < 10500) & (datos['Area'] > 70)
```

```python
datos_filtrados = datos[filtro1 | filtro2]
```

```python
datos_filtrados.head()
```

```python
datos.to_csv('datos_limpios.csv', sep=';', index=False)
```

```python
datos_filtrados.to_csv('datos_filtrados.csv', sep=';', index=False)
```

```python
datos['valor_mensual'] = datos['Valor'] + datos['Condominio']
```

```python
datos['valor_anual'] = datos['Impuesto'] + 12 * datos['valor_mensual']
```

```python
datos.head()
```

```python
datos.columns
```

```python
datos['Descripcion'] = (
    datos['Tipo'] + ' en ' + datos['Colonia'] +
    ' con ' + datos['Habitaciones'].astype(str) +
    ' habitaciones y ' + datos['Garages'].astype(str) + ' garages'
)
```

```python
datos['Tiene_suite'] = datos['Suites'] > 0
```

```python
datos['Tiene_suite'] = (datos['Suites'] > 0).map({True: 'Sí', False: 'No'})
```
